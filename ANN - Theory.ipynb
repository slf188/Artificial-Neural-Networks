# Activation Functions

## What are the commonly seen attributes inputs have?
* Weight: It implies how much weight to give the incoming input
* Bias: Offset value

Total Input:
z = weight * input + bias

Illustration:

activation_function.PNG

## What is the activation function?
Activation functions are mathematical equations that determine the output of a neural network, they are made in the nucleus of the neuron.

If we refer back to the previous total input then the activation function would look like: **f(z)**


## Which activation functions will we encounter?
1. Step Function: It outputs 0 or 1, useful for classification
2. Sigmoid Function: Curve like function

Formula:

sigmoid_function.PNG


3. Hyperbolic Tangent: tanh(z), outputs between -1 and 1
4. Rectified Linear Unit: max(0, z), good performance, good when dealing with **vanishing gradient**

Learn More: [Activation Functions](https://en.wikipedia.org/wiki/Activation_function)

# Multi-Class Classification Considerations

## How will Multi-Class Classification help us?
Activation Functions make sense for predicting a binary classification (either 0 or 1)
But when we encounter a multi-class classification then we better use this method
There are 2 types of situations:
* Non-Exclusive Cases: The data point can have multiple categories assigned

Example:

Photos can have multiple tags(family, beach, vacation, etc...)
* Mutually Exclusive Cases: Only one class per data point

Example:

Photos that contain animals(dog, cat)



## What activation functions should we use on each case?
Non-Exclusive Cases:
* Sigmoid Function

Mutually Exclusive Cases:
* Softmax Function: It calculates the probabilities of each target class over all possible target classes

# Cost Functions and Gradient Descent

## What is the purpose of the Cost Function and Gradient Descent?
We will use both to evaluate how far off our prediction is, i.e we will compare both the input and the output values

## What is the Cost Function?
AKA Loss Function, must be an average to measure how far off our prediction is from the true value.

We'll use the following variables:
* y: To represent the true value
* a: To represent neuron's prediction

Which in terms of weights and bias:
* w*x + b = z
* Pass z into an activation function 0(z) = a

We can think of the cost function as a function of four main things:

cost_elements.PNG

1. W: Neural Network's weight
2. B: Neural Network's biases
3. S^r: Input of the single the training sample
4. E^r: Desired output of that training sample

Formula:

cost_function.PNG

* a(x): Holds information about weights and biases

## What is Gradient Descent?
Let's imagine we have only one weight, and we need to figure out what value of w results in the minimum of C(w).

And if the cost function gets really complex then we need to use Gradient Descent.
We must calculate the slope at each point until it reaches zero, indicating a minimum

Illustration:

gradient_descent.PNG

The step size is known as the learning rate.
To adapt the step size in an efficient way, we must use **Adaptive Gradient Descent** known as [Adam](https://arxiv.org/abs/1412.6980), where the steps adapts to the situation.

# Backpropagation

## What is the purpose of using Backpropagation?
* Backpropagation starts at the last layer of the network

This means we already have:

z = w * a + b

a = 0(z)

C(...) = (a - y)^2

The main idea is: We can use the gradient to go back through the network and adust our weights and biases to minimize the output of the error vector on the last output layer.
