# Evaluating Performance - Classification Error Metrics

## Why Evaluating Performance?
*  Once our machine learning process is complete, we need to use performance metrics to evaluate how the model did

## Which Classification metrics do we need to understand?
*  Accuracy
*  Recall
*  Precision
*  F1-Score

## What is Accuracy?
*  Number of correct guesses divided by the total number of guesses
*  Useful when targeted classes are well balanced i.e the same amount of cat images as we have dog images
*  Not useful with unbalanced classes, if we had 99 cat images and 1 dog image, the model will get 99% accuracy which is false

## What is Recall?
*  Number of true positives divided by the sum of true positives and false negatives


## What is Precision?
*  Number of true positives divided by the sum of both true and negative positives


## What is F1-Score?
*  Combination of both Recall and Precision
*  Used when we want to find the mix of these metrics
* It rounds up values

# Unsupervised Learning

## What is Unsupervised Learning?
*  Algorithm using **unlabeled** data
*  There is no correct answer
*  Evaluation cannot be performed on this type of machine learning

## What tasks does Unsupervised Learning do?
*  Clustering
*  Anomaly Detection
*  Dimensionality Reduction

## What is Clustering?
*  The model groups similar data into clusters

Example: 

The model will group images where the animal has whiskers and another group where the animal has a tail

## What is Anomaly Detection?
*  Attempts to detect out of the ordinary pieces of data

Example:

Identifying fraudulent transactions

## What is Dimensionality Reduction?
*  The model reduces the number of features to better understand the dataset
